```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.preprocessing import normalize
#importing libraries
```

```python
# Sampling a subset of data
subset_size = 5000  # Adjust as needed
subset_data = main_anime_recommendation_factors.sample(subset_size, random_state=42)
```

```python
# Creating a sparse matrix for content based filtering
vectorizer = CountVectorizer()
content_matrix_sparse = vectorizer.fit_transform(subset_data['Genres'].fillna(''))
```

```python
# Calculate cosine similarity for content-based filtering with a sparse matrix
cosine_sim_content = cosine_similarity(content_matrix_sparse, dense_output=False)
```

```python
# Loading user ratings dataset
user_ratings_path = "anime_ratings.csv" 
```

```python
user_data = pd.read_csv(user_ratings_path, header=None, nrows=10000)
user_data[2] = pd.to_numeric(user_data[2], errors='coerce')  # Converting 'rating' to numeric
user_data = user_data.dropna(subset=[2])  # Droping rows with missing or non-numeric ratings
```

```python
numpy_user_data = user_data.to_numpy()
print(numpy_user_data)
#displaying user rating data
```

```python
print(user_data.head())
#displaying first few user rating data
```

```python
# Creating a user-item matrix for collaborative filtering
user_item_matrix = user_data.pivot(index=0, columns=1, values=2).fillna(0)
```

```python
# Calculating user similarity for collaborative filtering
user_similarity = 1 - pairwise_distances(user_item_matrix, metric='cosine')
```

```python
#Converting sparse matrix to dense matrix
dense_cosine_sim_content = cosine_sim_content.toarray()

# Resizing the content-based similarity matrix to match the collaborative similarity matrix
cosine_sim_content_resized = np.zeros(user_similarity.shape)

# taking the minimum dimensions to avoid shape mismatch
min_rows = min(cosine_sim_content.shape[0], user_similarity.shape[0])
min_cols = min(cosine_sim_content.shape[1], user_similarity.shape[1])

# Assigning the values to the resized matrix
cosine_sim_content_resized[:min_rows, :min_cols] = dense_cosine_sim_content[:min_rows, :min_cols]
```

```python
# Combining content-based and collaborative similarities for hybrid recommendations
hybrid_similarity = 0.5 * normalize(cosine_sim_content_resized, axis=1, norm='l2') + 0.5 * user_similarity
```

```python
# Getting hybrid recommendations from the given user
user_index = 30 #user id
hybrid_recommendations = list(enumerate(hybrid_similarity[user_index]))
hybrid_recommendations = sorted(hybrid_recommendations, key=lambda x: x[1], reverse=True)

```

```python
# Printing the top 10 hybrid recommendations
print("\nHybrid Recommendations:")
for idx, score in hybrid_recommendations[:10]:
    anime_id = user_item_matrix.columns[idx]
    anime_data_row = main_anime_recommendation_factors[main_anime_recommendation_factors['MAL_ID'] == anime_id].iloc[0]
    print(f"Anime ID: {anime_data_row['MAL_ID']}, Name: {anime_data_row['Name']}, Score: {anime_data_row['Score']}")
```

```python
synopsis_path = "anime_with_synopsis.csv"
synopsis_data = pd.read_csv(synopsis_path, header=None) 
```

```python
synopsis_headings = synopsis_data.iloc[0].values
#i am taking the first array of the dataset into headings
```

```python
synopsis_table = pd.DataFrame(synopsis_data.values[1:], columns=synopsis_headings)
synopsis_table
```

```python
# i am Loading the Universal Sentence Encoder which is a pre-trainded natural processing langauge
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
```

```python
# i am extracting synopsis data
synopsis_data = synopsis_table['sypnopsis'].tolist()
```

```python
# i am inserting a senario
user_input_senario = "The world of ninjas where a boy become the best ninja as creates a lots of friend through his adventure."
```

```python
# Combining user input senario and existing synopsis data
all_synopsis = [user_input_senario] + synopsis_data

# Ensuring that all elements are strings
all_synopsis = [str(item) for item in all_synopsis]
```

```python
# i am using Universal Sentence Encoder to encode the text into vectors
embeddings = embed(all_synopsis)
```

```python
# Calculating cosine similarity between the user input vector and all anime vectors
user_vector = embeddings[0].numpy().reshape(1, -1)
anime_vectors = embeddings[1:].numpy()
similarities = cosine_similarity(user_vector, anime_vectors)
```

```python
# Getting indices of anime with the highest similarity scores
recommended_anime_indices = np.argsort(similarities[0])[::-1][:10]
```

```python
# Printing the top 10 recommendations with similarity scores
print("\nTop 10 recommendations form the given scenarios:")
for idx in recommended_anime_indices:
    anime_data_row = synopsis_table.iloc[idx]
    similarity_score = similarities[0][idx]
    print(f"Anime ID: {anime_data_row['MAL_ID']}, Name: {anime_data_row['Name']}, Score: {anime_data_row['Score']}, Similarity Score: {similarity_score}")
```

```python
# Extracting similarity scores for the top 10 recommendations
similarity_scores = [similarities[0][idx] for idx in recommended_anime_indices]
```

```python
# Anime names for the x-axis
anime_names = [synopsis_table.iloc[idx]['Name'] for idx in recommended_anime_indices]
```

```python

# Creating a bar chart
plt.figure(figsize=(10, 6))
plt.bar(anime_names, similarity_scores, color='blue')
plt.xlabel('Anime Name')
plt.ylabel('Similarity Score')
plt.title('Similarity Scores for Top 10 Recommendations')
plt.xticks(rotation=45, ha='right')
plt.show()
```
